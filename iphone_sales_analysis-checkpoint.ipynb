# IPHONE SALES IN INDIA

##### The Indian smartphone market has shown a consistent preference for Apple iPhones, particularly models that offer a balanced mix of performance, price, and reliability. Based on data from Flipkart, we analyzed the top 10 highest-rated iPhones currently popular among Indian consumers. The following analysis reveals key patterns in consumer behavior and market demand.

```python
highest_rated=data.sort_values(by=["Star Rating"],ascending=False)
highest_rated=highest_rated.head(10)
print(highest_rated["Product Name"])
```

```python
print(highest_rated)
```

#### Now lets have a look at the number of ratings of the highest Rated iPhones on Flipkart

```python
import seaborn as sns
```

```python
ratings_per_iphone = highest_rated.groupby("Product Name")["Number Of Ratings"].sum().sort_values(ascending=False)

# Convert to DataFrame
ratings_df = ratings_per_iphone.reset_index()
ratings_df.columns = ["Product Name", "Number Of Ratings"]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(
    data=ratings_df,
    x="Product Name",
    y="Number Of Ratings",
    hue="Product Name",                 
    palette="coolwarm",
    dodge=False,                       
    legend=False                    
)

plt.title("Number of Ratings of Highest Rated iPhones")
plt.xlabel("Product Name")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### According to the above bar graph,APPLE iPhone 8 Plus has the most rating on Flipkart.Now lets have a look at the number of reviews of the highest rated iPhone on Flipkart.

```python
reviews_per_iphone = highest_rated.groupby("Product Name")["Number Of Ratings"].sum().sort_values(ascending=False)

# Convert to DataFrame
reviews_df = reviews_per_iphone.reset_index()
reviews_df.columns = ["Product Name", "Number Of Reviews"]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(
    data=reviews_df,
    x="Product Name",
    y="Number Of Reviews",
    hue="Product Name",                 # Fixes future warning
    palette="coolwarm",
    dodge=False,                        # Prevent side-by-side bars
    legend=False                        # Legend not needed
)

plt.title("Number of Reviews of Highest Rated iPhones")
plt.xlabel("Product Name")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Realtionship between (Sale Price and Number of Ratings) & (Discount Percentage and Numnber of Ratings)

```python
plt.figure(figsize=(10, 6))

# Scatter plot (bubble plot) with size mapped to 'Discount Percentage'
scatter = plt.scatter(
    data["Number Of Ratings"],
    data["Sale Price"],
    s=data["Discount Percentage"] * 10,   # Scale size for visibility
    c='royalblue',
    alpha=0.6,
    edgecolors='k'
)
sns.regplot(
    data=data,
    x="Number Of Ratings",
    y="Sale Price",
    scatter=False,
    color='darkblue',
    line_kws={"linewidth": 2}
)

plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.xlabel("Number of Ratings")
plt.ylabel("Sale Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

 ##### There is a negative linear relationship between the sale price of iPhone and the number of ratings.It means iPhones with lower sale peice are sold in India.Now lets have a look at the relationship between the discount percentage on iPhones on Flipkart and the number of ratings.

```python
plt.figure(figsize=(10, 6))

# Bubble scatter plot
plt.scatter(
    data["Discount Percentage"],
    data["Number Of Ratings"],
    s=100,  # Fixed bubble size or you can use data["Sale Price"] / 100 to scale
    alpha=0.6,
    c='mediumseagreen',
    edgecolors='k'
)

# Regression line
sns.regplot(
    data=data,
    x="Discount Percentage",
    y="Number Of Ratings",
    scatter=False,
    color='green',
    line_kws={"linewidth": 2}
)

# Titles and labels
plt.title("Relationship between Discount Percentage and Number of Ratings of iPhones")
plt.xlabel("Discount Percentage")
plt.ylabel("Number of Ratings")
plt.grid(True)
plt.tight_layout()
plt.show()
```
